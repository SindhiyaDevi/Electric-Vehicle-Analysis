```python
# Top Speed and Range filtration

df1 = df[df['Top_speed_kmph']>40] # speed more than 70KmH

plt.figure(figsize=(17,8))
plt.title("Speed and Range",fontsize=25)
plt.grid(axis='y')
plt.ylim(bottom=40,top=350)
plt.xticks(rotation = 90)
plt.xlabel("Model Name",fontsize=15)
plt.ylabel("Speed (KmH)",fontsize=15)
plt.bar(df1['Model'],df1['Top_speed_kmph'],label='Top Speed')
plt.plot(df1['Model'],df1['Range_km/charge'],label='Range',c='m')
plt.legend()
```

```python
# Filteration of models related to Range
df2 = df1[df1['Range_km/charge']>=100] # Range above 400 km

plt.figure(figsize=(15,5))
plt.title("Range_km/charge",fontsize=25)
plt.grid(axis='y')
plt.ylim(bottom=50,top=350)
plt.xticks(rotation = 90)
plt.xlabel("Model Name",fontsize=20)
plt.ylabel("Distance (Km)",fontsize=20)
plt.plot(df2['Model'],df2['Range_km/charge'],label='Range',c='r')
```

```python
#Filteration according to Price Range
df3 = df2[df2['Price']>40000] # cost greater than 5000
df4 =df3[df3['Price']<85000] # cost less than 80000

plt.figure(figsize=(15,7))
plt.title("Cost of vehicle",fontsize=15)
plt.xticks(rotation = 0)
plt.grid(axis='x')
plt.xlim(55000,100000)
plt.ylabel("Model Name",fontsize=12)
plt.xlabel("Price",fontsize=12)
plt.barh(df4['Model'],df4['Price'],label='Range')
plt.show()
```
