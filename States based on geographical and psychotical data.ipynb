```python
#k-means Clustering
# Importing Important Libraries
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
#clustering based on annual EV sale
df3= df[df['EV_Annual_sales_percentage_in_2022']>0]
df3.head()
```

```python
a = ["Working_people_in_percentage", "EV_Annual_sales_percentage_in_2022"]
n = df3.loc[:,a]
n.head()
```

```python
#ing the elbow method to find the optimal numbers of clusters

wcss= []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init="k-means++", random_state = 42)
    kmeans.fit(n)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss,color='blue', linestyle='dashed', marker='o')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

```python
#Training the K-means model on the dataset
kmeans= KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(n)
print(y_kmeans)
```

```python
n['cluster']=y_kmeans
n.head(20)
```

```python
#Visualising the clusters
n['cluster']=y_kmeans
sns.pairplot(data=n,hue='cluster',diag_kind=None)
```
